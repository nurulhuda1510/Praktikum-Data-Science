## Part 1: Box Plots

Box Plot adalah jenis visualisasi yang mampu memberikan insights tentang penyebaran data univariat numerik. Pada box plots, langsung terlihat **Q1, Q2 (Median), Q3, Interquartile Range, dan Outliers**.

### Contoh 1 

```python
fig = plt.figure()  
ax = sns.boxplot(data = df, x = 'age')
plt.show() 
```

Dari visualisasi di atas, dapat dikatakan bahwa:

- Q1 age berada di sekitar 35 
- Q2 (Median) age berada di 45 
- Q3 age berada di sekitar 60  
- Minimum (19) dan Maximum (over 90) masih berada di dalam Interquartile Range sehingga tidak ada Outliers

### Contoh 2

```python
fig = plt.figure()  
ax = sns.boxplot(data = df, x = 'weight')
plt.show() 
```

Dari visualisasi di atas, dapat dikatakan bahwa:

- Q1 Temperature berada di sekitar 150 
- Q2 (Median) Temperature berada di 170  
- Q3 Temperature berada di sekitar 200 
- Pada data diatas terdapat outlier dimana tinggi yang lebih dari 260 dianggap sebagai outlier

Kita juga bisa membuat boxplot berdasarkan 'kategori' dari feature lain. 

```python
fig = plt.figure()  
ax = sns.boxplot(data = df, x = 'glucose', y = 'diabetes')
plt.show() 
```

Dari visualisasi diatas dapat dilihat bahwa penderita diabetes nilai dari glucose milikinya lebih tinggi dibanding dengan orang yang tidak menderita diabetes

```python
fig = plt.figure()  
ax = sns.boxplot(data = df, x = 'diabetes', y = 'glucose')
plt.show() 
```

Dengan mengganti x dan y, kita bisa memutarkan grafik kita seperti di atas.

## Part 2: Univariate Visualization

## 1. Histogram

Histogram memberikan gambaran cepat tentang distribusi
suatu variabel. Dari histogram, kamu bisa menilai bentuk, pusat, dan penyebaran
distribusi. Histogram Termasuk kedalam Visualisasi univariat artinya melakukan visualisasi hanya pada 1 variabel saja. Biasanya, untuk data numerik, (data berwujud angka, bilangan)

```python
fig = plt.figure()  
ax = sns.histplot(data = df, x='glucose')
plt.show() 
```

Berdasarkan Visualisasi ini, dapat kita lihat pada data yang kita gunakan orang-orang dominan memiliki kadar glukosa kurang lebih 100

```python
fig = plt.figure()  
ax = sns.histplot(data = df, x='glucose', binwidth = 2)
plt.show() 
```

Pada histogram, suatu pengaturan yang dapat dilakukan adalah binwidth. Semakin besar binwidth maka semakin besar pula range dari setiap bar. Semakin kecil binwidth nya, maka detil dari penyebaran data akan semakin terlihat jelas.

## 2. Density Plot

Selain histogram yang kesannya 'patah-patah', ada juga 'Density Plot' yang memiliki fungsi yang sama, namun kali ini grafiknya lebih 'mulus'.

```python
fig = plt.figure()
ax = sns.kdeplot(data=df, x='glucose')
plt.show()
```

## Part 3: Bivariate Visualization

### Part 3.1. Continuous vs Continuous (Numerical vs Numerical)

Untuk visualisasi data Numerical vs Numerical, biasanya kita melakukan `scatterplot`.

```python
fig = plt.figure()  
ax = sns.scatterplot(data = df2, x='Area', y = 'Price')
plt.show()
```

Berdasarkan gambar di atas, kita dapat melihat bahwa terdapat kecenderungan bahwa semakin luas suatu Rumah, maka semakin mahal pula harga rumah tersebut

Kita juga dapat menambahkan warna di scatterplot. Pada contoh di atas, kita menambahkan warna. Warna lebih muda adalah rumah yang memiliki jumlah kamar lebih sedikit. Di sini, kita dapat melihat bahwa luas bertambah seiring bertambahnya jumlah kamar, dan hal ini juga diiringi dengan naiknya harga rumah tersebut

## Part 3.2. Categorical vs Categorical

Biasanya, ketika dua variabel yang ingin kita analisa sama-sama berbentuk kategorikal, kita ingin mengetahui jumlah dan proporsi antara dua variabel tersebut.

```python
fig = plt.figure()  
ax = sns.countplot(data = df, x='diabetes', hue = 'gender')
plt.show() 
```

Dari count plot di atas, dapat kita lihat bahwa penderita diabetes jauh lebih sedikit jika dibandingkan dengan jumlah penderita diabetes dan gender perempuan sedikit lebih banyak menderita penyakit diabetes

Bentuk lain dari categorical vs categorical visualization adalah Stacked Bar Plot. Pada Stacked Bar Plot, kita tidak memfokuskan pada jumlah, namun lebih menitikberatkan pada rasio.

```python
pd.crosstab(df['diabetes'], df['gender']).apply(lambda x: x*100/x.sum(), axis = 1)
```

```python
pd.crosstab(df['diabetes'], df['gender']).apply(lambda x: x*100/x.sum(), 
axis = 1).plot(kind = 'bar', stacked = True)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left', title='Weather Type')
plt.show()
```

Begini interpretasinya:

- 57% dari Gender female adalah penderita diabetes, 43% dari Gender male adalah penderita diabetes
- 59% dari Gender female bukan penderita diabetes, 41% dari Gender male bukan penderita diabetes

Bagaimana Jika Sebaliknya 

```python
pd.crosstab(df['gender'], df['diabetes']).apply(lambda x: x*100/x.sum(), 
axis = 1)
```

```python
pd.crosstab(df['gender'], df['diabetes']).apply(lambda x: x*100/x.sum(), 
axis = 1).plot(kind = 'bar', stacked = True)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left', title='Weather Type')
plt.show()
```

- 15% dari gender female mengalami diabetes, 85% lainnya bukan penderita diabetes
- 16% dari gender male adalah penderita diabetes, 84% lainnya bukan penderita diabetes

## Part 3.3. Categorical vs Continuous

Pada visualisasi Categorical vs Continuous, biasanya kita ingin melihat distribusi dari suatu variabel univariat namun ingin dipecah berdasarkan 'kelas-kelas' tertentu (pemecahan ini tergantung variabel kategorikal).

```python
fig = plt.figure()  
ax = sns.histplot(data = df, x='gender', hue = 'diabetes', binwidth = 5)
plt.show()
```

Sebagai contoh, kita bisa me 'numpukkan' dua histogram dalam satu grafik, namun menyajikan mereka dalam warna yang berbeda. Dalam grafik ini terlihat bahwa jumlah penderita diabetes jauh lebih sedikit dibandingkan dengan yang bukan penderita diabetes baik dari gender pria dan wanita

## Part 4: Time Series Visualization

```python
# importing packages
import seaborn as sns
import pandas as pd
  
# creating data
df = pd.DataFrame({'Date': ['2019-10-01', '2019-11-01', 
                            '2019-12-01','2020-01-01', 
                            '2020-02-01', '2020-03-01',
                            '2020-04-01', '2020-05-01', 
                            '2020-06-01'],
                     
                   'Col_1': [34, 43, 14, 15, 15,
                             14, 31, 25, 62],
                     
                   'Col_2': [52, 66, 78, 15, 15,
                             5, 25, 25, 86],
                     
                   'Col_3': [13, 73, 82, 58, 52,
                             87, 26, 5, 56],
                   'Col_4': [44, 75, 26, 15, 15,
                             14, 54, 25, 24]})
  
# create the time series plot
sns.lineplot(x = "Date", y = "Col_1", data = df)
sns.lineplot(x = "Date", y = "Col_2", data = df)
sns.lineplot(x = "Date", y = "Col_3", data = df)
plt.ylabel("Col_1 and Col_2")
plt.xticks(rotation = 25)

plt.show()
```

```python
# Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set figure default figure size
plt.rcParams["figure.figsize"] = (10, 6)
# Create a random number generator for reproducibility
rng = np.random.default_rng(1111)

# Get some random points!
x = np.array(range(10))
y = rng.integers(10, 100, 10)
z = y + rng.integers(5, 20, 10)
```

```python
plt.plot(
    x, z, ls="--", lw=3, 
    marker="X", markersize=10, markerfacecolor="red", markeredgecolor="black",
    label="Income"
)
plt.plot(
    x, y, ls=":", 
    marker="o", markersize=15, markerfacecolor="None", 
    label="Expenses"
)
plt.legend()
plt.show()
```

## Pie Chart

Pie chart digunakan Menampilkan Proporsi atau Persentase Kategori dalam Satu Keseluruhan

```python
# library
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,5)
# create random data
names='groupA', 'groupB', 'groupC', 'groupD',
values=[12,11,3,30]
 
# Create a set of colors
colors = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897']

# Use it thanks to the color argument
plt.pie(values, labels=names, labeldistance=1.15, wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }, colors=colors);
```

```python
data = df['diabetes'].value_counts()

# Buat pie chart
plt.figure(figsize=(6,6))
plt.pie(data, labels=data.index, autopct='%1.1f%%')
plt.title("Distribusi Diabetes")
plt.show()
```
